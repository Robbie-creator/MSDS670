```python
#importing Required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_excel("Billionaires_Analysis_2021.xlsx")
df.head()
```

# visualization 

## 1.Distribution of Billionaire Net Worth (Histogram)

```python
net_worth_billions = df["total_net_worth"] / 1e9  # (1e9 =1 Billion) Good for visualization

plt.figure()
plt.hist(net_worth_billions, bins=20)
plt.title("Distribution of Billionaire Net Worth (2021)")
plt.xlabel("Net Worth (Billions USD)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

## 2. Average Net Worth by Industry 

```python
industry_avg = df.groupby("industry")["total_net_worth"].mean() / 1e9

# Sort in descending order
industry_avg = industry_avg.sort_values(ascending=False)

plt.figure()
plt.barh(industry_avg.index, industry_avg.values)
plt.title("Average Net Worth by Industry")
plt.xlabel("Average Net Worth (Billions USD)")
plt.ylabel("Industry")
plt.tight_layout()
plt.show()
```

# 3. Net Worth vs YTD change

```python
net_worth = df["total_net_worth"] / 1e9
ytd_change = df["ytd_change"] / 1e9

plt.figure()
plt.scatter(net_worth, ytd_change)
plt.title("Net Worth vs YTD Change (2021)")
plt.xlabel("Total Net Worth (Billions USD)")
plt.ylabel("YTD Change (Billions USD)")
plt.tight_layout()
plt.show()
```

# 4. Top 5 Billionaire in each sector

```python
industries = df["industry"].unique()

for industry in industries:
    
    industry_df = df[df["industry"] == industry]
    top5 = industry_df.sort_values("total_net_worth", ascending=False).head(5)
    
    # Convert to billions
    net_worth_billions = top5["total_net_worth"] / 1e9

    plt.figure()
    plt.bar(top5["name"], net_worth_billions)
    plt.title(f"Top 5 Billionaires in {industry} Industry (2021)")
    plt.xlabel("Billionaire")
    plt.ylabel("Total Net Worth (Billions USD)")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
